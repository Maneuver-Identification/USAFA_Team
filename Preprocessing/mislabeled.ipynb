## Mislabeled Sorties

This code iterates through each of the png and tsv folders sorted into good and bad and makes sure there are no discrepancies between the two. These were manually sorted, so there is a case in which a png file marked as 'bad' has a corresponding tsv file marked as 'good', which is pointed out in this code.

```python
import sys, os
rootdir = sys.argv[1] #should navigate to a directory that has 4 folders: tsv-good, tsv-bad, png-good, and png-bad

badPNG = set()
goodPNG = set()
badTSV = set()
goodTSV = set()

for subdir, dirs, files in os.walk(rootdir):

        if ("bad" in subdir):
            #bad TSVs
            if ("tsv" in subdir):
                for file in files:
                    badTSV.add(file.split(".")[0])
            elif ("png" in subdir):
                for file in files:
                    badPNG.add(file.split(".")[0])
        else: #good in subdir:
            # good TSVs
            if ("tsv" in subdir):
                for file in files:
                    goodTSV.add(file.split(".")[0])
            elif ("png" in subdir):
                for file in files:
                    goodPNG.add(file.split(".")[0])

#Finds Descrepancies
mismatched_good = set(set(goodPNG) ^ set(goodTSV))
mismatched_bad = set(set(badPNG) ^ set(badTSV))
#Consolidates Descrepancies
temp_missed = mismatched_bad - mismatched_bad
misidenfitied = list(mismatched_good) + list(temp_missed)
#displays Descrepancies
print(len(misidenfitied), "sorties have been identified as mislabeled")
print(misidenfitied)
```
